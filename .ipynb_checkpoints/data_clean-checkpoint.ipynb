```python
df = pd.read_stata("/users/quasar/downloads/10sd_jan15_2014/10SD_jan15.dta")
df
```

```python
df.set_index(['Country','Year'])
```

```python
df.loc[(df.Country == 'BRA') & (df.Year == 2005)]
```

```python
# VA = value added , EMP = employment 
# df.loc[df["Country"].isin(["BRA", "IND", "MEX", "USA" ]), ["Country", "Variable", "AGR"]].loc[df.Year == 2005]
df_clean = df.loc[df["Country"].isin(["BRA", "IND", "MEX", "USA" ])]
df_clean
```

```python
df_emp = df_clean.loc[df.Variable == 'EMP']
df_emp
```

```python
df_emp['agri'] = df_emp.AGR
df_emp['manufacturing'] = df_emp.MIN + df_emp.MAN
df_emp['service'] = df_emp.PU + df_emp.CON + df_emp.WRT + df_emp.TRA + df_emp.FIRE + df_emp.GOV + df_emp.OTH
df_emp['agri_share'] = df_emp.AGR / df_emp.SUM
df_emp['manu_share'] = df_emp.manufacturing / df_emp.SUM
df_emp['service_share'] = df_emp.service / df_emp.SUM
```

```python
df_emp.loc[df_emp['Country'] == 'USA']
df_emp.loc[df_emp["Country"].isin(["BRA", "IND", "MEX", "USA" ]), ["Year","Country", "agri_share", "manu_share", "service_share"]].loc[df.Year == (2000)]
```

```python
df_emp.loc[(df_emp['Country'] == 'IND') & (df_emp['Year'] >= 1980)]['agri_share']
df_ind = df_emp.loc[(df_emp['Country'] == 'IND') & (df_emp['Year'] >= 1980) & (df_emp['Year'] <= 2005), ['Year', 'agri_share']]
df_mex = df_emp.loc[(df_emp['Country'] == 'MEX') & (df_emp['Year'] >= 1960) & (df_emp['Year'] <= 2000), ['Year', 'agri_share']]
df_bra = df_emp.loc[(df_emp['Country'] == 'BRA') & (df_emp['Year'] >= 1960) & (df_emp['Year'] <= 2000), ['Year', 'agri_share']]
df_USA = df_emp.loc[(df_emp['Country'] == 'USA') & (df_emp['Year'] >= 1960) & (df_emp['Year'] <= 2000), ['Year', 'agri_share']]
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
# use human capital ratio from table 1 to calculate values below 
df_emp['agri_eff'] = df_emp.AGR
df_emp['manufacturing_eff'] = (df_emp.MIN + df_emp.MAN) * 2.56
df_emp['service_eff'] = (df_emp.PU + df_emp.CON + df_emp.WRT + df_emp.TRA + df_emp.FIRE + df_emp.GOV + df_emp.OTH) * 4.17 
df_emp['total_eff'] = df_emp['agri_eff'] + df_emp['manufacturing_eff'] + df_emp['service_eff']
df_emp['agri_share_eff'] = df_emp.AGR / df_emp.total_eff
df_emp['manu_share_eff'] = df_emp.manufacturing / df_emp.total_eff
df_emp['service_share_eff'] = df_emp.service / df_emp.total_eff
```

```python
df_ind_eff = df_emp.loc[(df_emp['Country'] == 'IND') & (df_emp['Year'] >= 1980) & (df_emp['Year'] <= 2005), ['Year', 'agri_share_eff']]
df_mex_eff = df_emp.loc[(df_emp['Country'] == 'MEX') & (df_emp['Year'] >= 1960) & (df_emp['Year'] <= 2000), ['Year', 'agri_share_eff']]
df_bra_eff = df_emp.loc[(df_emp['Country'] == 'BRA') & (df_emp['Year'] >= 1960) & (df_emp['Year'] <= 2000), ['Year', 'agri_share_eff']]
df_USA_eff = df_emp.loc[(df_emp['Country'] == 'USA') & (df_emp['Year'] >= 1960) & (df_emp['Year'] <= 2000), ['Year', 'agri_share_eff']]
```

```python
# Calibration values for all countries 
ca_bar = 1552.7
cs_bar = 879.6
rho = 0.5
# For india 
n_a = df_ind_eff.agri_share_eff
lambda_a = 0.1
lambda_m = 0.15
lambda_s = 0.75
tau_m = 0.24
tau_s = 0.13
```
